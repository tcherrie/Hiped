# Simple domains examples

```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import sys, os
sys.path.append(os.path.relpath("../../../src/python"))
import hiped as hp
```

## 1) 0D domain

```python
domain0 = hp.Domain(1) # a single dot
plt.clf()
domain0.plot()
```

## 2) 1D domain

```python
domain1 = hp.Domain(2) # a 1D line segment has 2 vertices
plt.figure()
domain1.plot()
```

## 3) 2D domain

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def f(x):
    plt.clf()
    hp.Domain(x).plot()
    plt.show()

plt.figure()
_ = interact(lambda x : f(x), x= widgets.IntSlider(min=3, max=20, step=1, value=5))
```

## 4) 3D domain

```python
plt.figure()
_ = interact(lambda x : f(x), x = ["tetra", "cube", "diamond3", "diamond6", "prism3", "prism6"])
```
